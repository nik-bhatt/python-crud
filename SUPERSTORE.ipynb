```python
# prompt: index and value pie chart

import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame({'Index': dataset['Ship Mode'].unique(), 'Value': dataset['Ship Mode'].value_counts()})

plt.figure(figsize=(5, 4))
plt.pie(df['Value'], labels=df['Index'], autopct="%1.1f%%", startangle=90)
plt.title('Ship Mode')
plt.axis('equal')
plt.show()
```
